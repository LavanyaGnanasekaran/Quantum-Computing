```python
import time
from qiskit import ClassicalRegister, QuantumRegister
from qiskit import QuantumCircuit
from qiskit import execute
from qiskit import IBMQ
import math
import operator
from qiskit import Aer
from GenerateLargePrime import generateLargePrime
from QFTAddition import add
from QFTMultiplication import mul
# generate random integer values
from random import seed
from random import randint
from Mont_Modular_Multiplication import Mont_Mul
from Mont_Modular_Exponentiation import Mont_Exp
from Binary_Extended_GCD import Mod_Inv

# when you restart the Jupyter kernel, you need to uncomment the following two lines
# provider = IBMQ.enable_account('<your-token>')
# qasm_sim = provider.get_backend('ibmq_qasm_simulator')        
        
# P and Q random prime number Generation, change the bit size in (generateLargePrime) function if you need to change
keysize = input("Enter the keysize")
keysize = int(keysize)
P = bin(generateLargePrime(keysize))[2:]
Q = bin(generateLargePrime(keysize))[2:]

while (P == Q):
    Q = bin(generateLargePrime(keysize))[2:]

print("P = ",int(P,2))
print("Q = ",int(Q,2))

pminus1 = int(P,2) - 1
qminus1 = int(Q,2) - 1
# print(pminus1, qminus1)

pminus1 = format(pminus1, "b")
qminus1 = format(qminus1, "b")
# print(pminus1, qminus1)

# Product of P and Q
N = mul(P,Q)
print("N = ", N) 

# Product of P-1 and Q-1
PhiofN = mul(pminus1, qminus1)

print("PhiofN =", PhiofN)

# seed random number generator
seed(1)
e = str(bin(randint(1, (int(PhiofN,2) - 1)))[2:])
print('e = ', e)

d = Mod_Inv(PhiofN,e)
print("d = ", d)

# Getting message from User
M = input("Enter the message in binary")
L = len(M)
start_time_encryption = time.time()
# Encryption process or calculating cipher text
C = Mont_Exp(M, e, N)
print('Encrypted text: ', C)
print("--- %s seconds ---" % (time.time() - start_time_encryption))

C = C[(-L):]
start_time_decryption = time.time()
# Decryption process or calculating original text
M = Mont_Exp(C, d, N)
print('Original Text: ', M)
print("--- %s seconds ---" % (time.time() - start_time_decryption))
```
